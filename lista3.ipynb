```
print("Questão 2, item b):")
print("Escolhemos o algoritmo de classificação do sklearn pois é de fácil utilização.")
# O estimador de decisão tem um atributo chamado tree_ que guarda a estrutura
# inteira da árvore e permite acesso a atributos de baixo nível. A árvore binária
# tree_ é representada como um numero de arrays paralelos. O i-ésimo elemento de cada
# array contém informação sobre o nó "i". O nó 0 é a raiz da árvore.
#
# Dentre estes arrays, temos:
#   - left_child, id do filho esquerdo do nó
#   - right_child, id do filho direito do nó
#   - feature, característica usada para dividir o nó
#   - threshold, valor limite do nó
#
# A estrutura da árvore pode ser atravessada para computar várias propriedades como
# a profunididade de cada nó e se ele é ou não uma folha.
#
# O classificador utiliza a impuridade GINI como critério de escolha e usa a estratégia
# de melhor divisão para dividir os nós em seus filhos pelos melhores atributos.
#
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
tree.plot_tree(clf)
```

```
print("Questão 2, item d):")
print("Avaliando o desempenho manualmente")
acertos = 0
for index, expectedResult in enumerate(y_test):
  if list(clf.predict(X_test))[index] == expectedResult:
    acertos += 1

print("Acertos: ",acertos)
print("Tamanho do teste: ", len(y_test))
print("Porcentagem de acerto: ",(acertos/len(y_test))*100,"%")
print("Teste de Acurácia do sklearn: ",accuracy_score(y_test, clf.predict(X_test)))

print("Há um falso negativo e dois falsos positivos, como podemos ver na matriz de confusão não-normalizada.")

titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(clf, X_test, y_test,
                                 display_labels=['versicolor', 'virginica'],
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
