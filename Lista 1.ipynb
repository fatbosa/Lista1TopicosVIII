```
import numpy as np
import pandas as pd
import statistics
from collections import Counter
import pylab
import scipy.stats
import statsmodels.api as sm
import seaborn as sn
import matplotlib.pyplot as plt



print("Alunos: Gabriel Barbosa Menezes Dantas e Lucas Matheus Carvalho de Assis")
print("Este bloco inicializa as bibliotecas utilizadas na atividade.")
```

```
print("Questão A:")

print("Gerando os 1000 valores com distribuição normal sendo média 0 e desvio padrão 1.")
X = np.random.normal(0, 1, 1000)
Y = np.random.normal(0, 1, 1000)
```

```
print("Questão A, Item 1): Calculando a média, mediana, moda, variância e desvio padrão de X:")
print("Média de X: ",np.mean(X))
print("Mediana de X: ",np.median(X))
#este if/else verifica se há moda no conjunto de dados
if (Counter(X).most_common(1)[0][1]) == 1:
  print("Todos os valores aparecem uma vez cada, logo não há moda.")
else:
  print("Moda de X: ",Counter(X).most_common(1)[0][0],", se repete ",Counter(X).most_common(1)[0][1]," vezes")
print("Variância de X: ",np.var(X))
print("Desvio Padrão de X: ",np.std(X))
```

```
print("Questão A, Item 1): Calculando a média, mediana, moda, variância e desvio padrão de Y:")
print("Média de Y: ",np.mean(Y))
print("Mediana de Y: ",np.median(Y))
#este if/else verifica se há moda no conjunto de dados
if (Counter(Y).most_common(1)[0][1]) == 1:
  print("Todos os valores aparecem uma vez cada, logo não há moda.")
else:
  print("Moda de Y: ",Counter(Y).most_common(1)[0][0],", se repete ",Counter(Y).most_common(1)[0][1]," vezes")
print("Variância de Y: ",np.var(Y))
print("Desvio Padrão de Y: ",np.std(Y))
```

```
print("Questão A, Item 2): Calculando os histogramas de X e Y:")
histx = pd.DataFrame({'X':X})
histx.hist()
histy = pd.DataFrame({'Y':Y})
histy.hist()
print("Podemos ver nos resultados abaixo que a média deve sempre estar no intervalo")
print("de valores com mais integrantes na distribuição normal, devido ao formato da")
print("curva da função. Podemos observar também que a mediana nem")
print("sempre está neste intervalo, pois sua posição pode variar, dependendo da")
print("população de valores antes ou depois do meio da curva.")
print("Por causa do grande intervalo de números, é comum obter um conjunto de valores")
print("sem moda, apesar de que se houvesse, a moda provavelmente se encontraria no centro.")
print("A variância se mostra no histograma no fato de que a curva é mais densa")
print("nos intervalos medianos, já o desvio padrão é próximo ao que usamos para gerar os dados")
print("e isso é visível no histograma pois os intervalos no meio da curva somados ou")
print("subtraídos do desvio padrão são os mais populados.")
print("Histograma de X e y:")
```

```
print("Questão A, Item 3): Calculando a assimetria e a curtose de X:")
print("Assimetria de X:",scipy.stats.skew(X))
print("Curtose de X:",scipy.stats.kurtosis(X))
```

```
print("Questão A, Item 3): Calculando a assimetria e a curtose de Y:")
print("Assimetria de Y:",scipy.stats.skew(Y))
print("Curtose de Y:",scipy.stats.kurtosis(Y))
```

```
print("Questão A, Item 4): Plot do gráfico Q-Q de X:")
sm.qqplot(X, line ='45')
#no cálculo da normalidade utilizamos o teste de jarque bera:
print("Normalidade de X, quanto maior que zero, menos próxima da distribuição normal: ",sm.stats.stattools.jarque_bera(X)[0])
pylab.show()
```

```
print("Questão A, Item 4): Plot do gráfico Q-Q de Y:")
sm.qqplot(Y, line ='45')
#no cálculo da normalidade utilizamos o teste de jarque bera:
print("Normalidade de Y, quanto maior que zero, menos próxima da distribuição normal:: ",sm.stats.stattools.jarque_bera(Y)[0])
pylab.show()
```

```
print("Questão A, Item 5): Matrizes de Covariância e Correlação entre X e Y:")
print("Matriz de Covariância entre X e Y: ")
sn.heatmap(np.cov(X,Y), annot=True)
plt.show()
data = {'X': X,
        'Y': Y
        }

df = pd.DataFrame(data,columns=['X','Y'])

corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
print("Matriz de Correlação entre X e Y: ")
plt.show()
print("A matriz de covariância mostra a interdependência entre as variáveis X e Y,")
print("enquanto a matriz de correlação mostra a relação estatística entre as duas.")
print("A covariância ajuda a determinar como as duas variáveis se relacionam enquanto")
print("a correlação ajuda a determinar o quanto elas se relacionam.")
print("Aqui podemos ver então que X e Y não tem muita relação entre si, apenas coincidental.")
```

```
print("Questão A, Item 6): Se duas variáveis possuem correlação 1, se movem na mesma")
print("direção e proporção. Se duas variáveis possuem correlação igual a 0, não possuem")
print("relação entre si. Se sua correlação é -1, se movem em direções opostas, mas na")
print("mesma proporção.")
```
